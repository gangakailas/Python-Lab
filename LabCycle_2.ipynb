```python
list1=[1, 1]

def get_rabbit_pair(n):
  for i in range(0, n):
    print("|   ",i+1,"         |     ",list1[i],"            |")
    print("-"*38)
    sum=list1[i]+list1[i+1]
    list1.append(sum)


x=int(input("Enter no. of months: "))
print("-"*38)
print("|  MONTHS       |   NO. OF PAIRS     |")
print("-"*38)
get_rabbit_pair(x) 
```

```python
#converting space separated string to a list
str1=input("Enter a some space separated numbers= ")
list1=str1.split()
print(list1)

#rotating the list element by 'k' position
k=int(input("Enter the position by which the list elements are to be rotated= "))
list1=list1[k:]+list1[0:k]
print(list1)
print("\n")

#converting the list to a tuple using comprehension
print("After list to tuple conversion:")
tup1=tuple(i for i in list1)
print(tup1)
print("\n")

#removing duplicate values from tuple and turning it back to a list
tup1=tuple(set(tup1))
print("Tuple after duplicate removal:")
print(tup1)
print("Conversion back to list:")
mylist=list(tup1)
print(mylist)
print("\n")

#creating a new list using function f(x)=x^2-x 
newlist=[]
mylist1=[]
for i in mylist:           
  mylist1.append(int(i))     #for converting the elements in list to int type(from str type)
for i in mylist1:           
  fnvalue=(i*i)-i
  newlist.append(fnvalue)   #creating list according to function
print("Creation of new list using function f(x)=x^2-x ")
print("Source list= ", mylist1)
print("Created list= ", newlist)
print("\n")

#sorting and merging mylist1 and newlist
mylist1.sort()
print("First list after sorting: ")
print(mylist1)

newlist.sort()
print("Second list after sorting: ")
print(newlist)

finallist=mylist1+newlist
finallist.sort()
print("Final sorted list after merging:")
print(finallist)
```

```python
import json
def create_list(way):
  file1=open(way,"r")
  result=file1.read()
  file1.close()
  return result

def list_of_dict(list1):
  dummy=json.loads(list1)
  return (dummy)

def setosa(dlist):
  print("Details of all Setosa Flowers:")
  for i in dlist:
    if i["species"]=="setosa":
      print(i)


def max_min(dlist):
  species=[]
  
  for i in dlist:
    species.append(i["species"])

  species=list(set(species))
  sepal_Area=[]
  petal_Area=[]

  for i in species:
    for j in dlist:
      if j["species"]==i:
        sepal_Area.append(j["sepalLength"]*j["sepalWidth"])
        petal_Area.append(j["petalLength"]*j["petalWidth"])
    
    print("Minimum petal area of",i,"=",round(min(petal_Area),2))
    print("Maximum sepal area of",i,"=",round(max(sepal_Area),2))

def total_area(dlist):
  for i in dlist:
    p_area=i["petalLength"]*i["petalWidth"]
    s_area=i["sepalLength"]*i["sepalWidth"]
    total_area=p_area+s_area
    i.update({"TotalArea":round(total_area,2)})

  sortedlist=sorted(dlist, key=lambda i:i["TotalArea"])
  for i in sortedlist:
    print(i)


way= "/content/drive/MyDrive/iris.json"
list1=create_list(way)
print("List with each line of file as elements:")
print(list1,"\n")

dlist=list_of_dict(list1)
print("List of Dictionary objects:")
print(dlist,"\n")
setosa(dlist)
print("\n")
max_min(dlist)
print("\n")
print("Sorted according to Total Area")
total_area(dlist)
```

```python
class Box:
  def __init__(self, length, breadth=None, height= None):
    self.length=length
    self.breadth=breadth
    self.height=height
    self.area=0
    self.volume=0

    if breadth is None and height is None: #cube
      self.breadth=length
      self.height=length
    elif height is None:                   #square prism
      self.height=length
    else:                                  #rectangular prism
      pass

  def display(self):
    print("Length= ",self.length)
    print("Breadth= ",self.breadth)
    print("Height= ",self.height)

  def find_area(self):
    self.area=2*(self.length*self.breadth+self.length*self.height+self.breadth*self.height)

  def find_volume(self):
    self.volume=self.length*self.breadth*self.height

N=int(input("Enter the number of boxes you prefer: "))
box=[]
for i in range(N):
  print("1.cube")
  print("2.square prism")
  print("3.rectangular prism")
  choice=int(input("Enter your choice: "))
  if(choice==1):
    l=int(input("Enter a dimension: "))
    box.append(l)
  elif(choice==2):
    l=int(input("Enter the length: "))
    b=int(input("Enter the breadth: "))
    box.append((l,h))
  elif(choice==3):
    l=int(input("Enter the length: "))
    b=int(input("Enter the breadth: "))
    h=int(input("Enter the height: "))
    box.append((l,b,h))

print(box)
l=[]
b=[]
h=[]
a=[]
v=[]
ratios=[]
for i in box:
  if type(i)==tuple and len(i)==3:
    shape=Box(i[0],i[1],i[2])
  elif type(i)==tuple and len(i)==2:
    shape=Box(i[0],i[1])
  elif type(i)==int:
    shape=Box(i)
  shape.display()                   #to print length, breadth, height
  
  l.append(shape.length)            #to store details in order
  b.append(shape.breadth)
  h.append(shape.height)

  shape.find_area()                 #finding, printing and storing area value
  area=shape.area
  print("Area= ",area)
  a.append(area)                   

  shape.find_volume()               #finding, printing and storing volume value
  volume=shape.volume
  print("Volume= ",volume)
  v.append(volume)

  ratios.append(volume/area)

print(ratios)
print("Details of the box with highest Volume to Area Ratio")
count=0
for i in ratios:
  if i==max(ratios):
    print("Ratio= ",i)
    print("Length= ", l[count])
    print("Breadth= ", b[count])
    print("Height= ", h[count])
    print("Volume= ", v[count])
    print("Area= ", a[count])
  count=count+1
```

```python
import math

class Shapes:

  def __init__(self, r, h=None):
    self.radius=r
    self.height=h


  def printVolume(self):
    pass

  def printArea(self):
    pass

class Cylinder(Shapes):

  def printVolume(self):
    volume=(math.pi*(self.radius**2))*self.height
    print("Cylinder Volume: ",round(volume,2))

  def printArea(self):
    area=2*math.pi*(self.radius**2)+2*math.pi*self.radius*self.height 
    print("Cylinder Area: ",round(area,2))

class Sphere(Shapes):

  def printVolume(self):
    volume=(4/3)*math.pi*(self.radius**3)
    print("Sphere Volume: ",round(volume,2))

  def printArea(self):
    area=4*math.pi*(self.radius**2)
    print("Sphere Area: ",round(area,2))

c_radius=int(input("Enter radius for Cylinder: "))
c_height=int(input("Enter height for Cylinder: "))
cylinder = Cylinder(c_radius, c_height)
cylinder.printVolume()
cylinder.printArea()

s_radius=int(input("Enter radius for Sphere: "))
sphere = Sphere(s_radius)
sphere.printVolume()
sphere.printArea()
```
