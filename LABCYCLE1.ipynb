```python
  #finding sum, reverse, difference between products of digits at odd and even positions
def sum_of_digits(x):
  sum=0
  while x>0:
    r=x%10
    sum=sum+r
    x=x//10
  return sum

def reverse_of_number(y):
  rev=0
  while y>0:
    r=y%10
    rev=(rev*10)+r
    y=y//10
  return rev

def dif_odd_even(z):
  evenprod=1
  oddprod=1
  while z>0:
    r=z%10
    oddprod=oddprod*r
    z=z//10
    
    r=z%10
    evenprod=evenprod*r
    z=z//10
  
  dif=oddprod-evenprod
  return dif

num=int(input("Enter a 4-digit number of your choice= "))
sum=sum_of_digits(num)
reverse=reverse_of_number(num)
difference=dif_odd_even(num)
print("The sum of the digits in the number= ",sum)
print("The reverse of the number= ",reverse)
print("The difference between product of digits at odd and even positions= ",difference)
```

```python
#reading sides of two triangles and finding the area percentage
import math
def read_sides():
  x1=int(input("Enter side 1 of triangle1= "))
  y1=int(input("Enter side 2 of triangle1= "))
  z1=int(input("Enter side 3 of triangle1= "))

  x2=int(input("Enter side 1 of triangle2= "))
  y2=int(input("Enter side 2 of triangle2= "))
  z2=int(input("Enter side 3 of triangle2= "))

  return x1, y1, z1, x2, y2, z2

def find_area(p,q,r):
  s=(p+q+r)/2
  area=math.sqrt(s*(s-p)*(s-q)*(s-r))
  return area

a1,b1,c1,a2,b2,c2=read_sides()
area_tri1=find_area(a1,b1,c1)
area_tri2=find_area(a2,b2,c2)

print("Area of triangle1=", area_tri1)
print("Area of triangle2=", area_tri2)

sum=area_tri1+area_tri2
print("Sum of the area of the triangles= ",sum)

print("Triangle1 area contribution= ", (area_tri1/sum)*100)
print("Triangle2 area contribution= ", (area_tri2/sum)*100)


```

```python
#generating payslip
def g_salary(bp,da,hra,ma):
  gs=bp+da+hra+ma
  return gs

def deduction(pt,pf,it):
  d=pt+pf+it
  return d

def n_salary(gs,d):
  ns=gs-d
  return ns

def payslip(empname, code, bpay):
  if bpay<10000:
    da=(bpay*5)/100
    hra=(bpay*2.5)/100
    ma=500
    pt=20
    pf=(bpay*8)/100
    it=0
  elif bpay>=10000 and bpay<30000:
    da=(bpay*7.5)/100
    hra=(bpay*5)/100
    ma=2500
    pt=60
    pf=(bpay*8)/100
    it=0
  elif bpay>=30000 and bpay<50000:
    da=(bpay*11)/100
    hra=(bpay*7.5)/100
    ma=5000
    pt=60
    pf=(bpay*11)/100
    it=11
  else:
    da=(bpay*25)/100
    hra=(bpay*11)/100
    ma=7000
    pt=80
    pf=(bpay*12)/100
    it=20
   

  gross_salary=g_salary(bpay,da,hra,ma)
  deduct=deduction(pt,pf,it)
  net_salary=n_salary(gross_salary,deduct)

  print("********* PAYSLIP *********")
  print("Employee Name= ",empname)
  print("Employee Code= ",code)
  print("Employee Basic Pay= ",bpay)
  print("Employee DA= ",da)
  print("Employee HRA= ",hra)
  print("Employee MA= ",ma)
  print("Employee PT= ",pt)
  print("Employee PF= ",pf)
  print("Employee IT= ",it)
  print("Employee Gross Salary= ",gross_salary)
  print("Employee Deduction= ",deduct)
  print("Employee Net Salary= ",net_salary)


employee_name=str(input("Enter the Employee Name= "))
code=int(input("Enter your Code= "))
basic_pay=int(input("Enter your Basic Pay= "))
payslip(employee_name, code, basic_pay)
```

```python
#string operations
def substrings(x):
  print("All possible substrings:")
  l=len(x)
  for i in range(0,l+1):
    for j in range(i+1,l+1):
      print(x[i:j],end=',')
  print(" ")
      

def substring_with_length(x,y):
  print("All possible substrings with length",y,":")
  l=len(x)
  for i in range(0,l+1):
    for j in range(i+1,l+1):
      if len(x[i:j])==y:
        print(x[i:j],end=',')
  print(" ")


def substring_length_distchar(x,y,z):
  print("All possible substrings with length",y,"and",z,"distinct characters:")
  l=len(x)
  for i in range(0,l+1):
    for j in range(i+1,l+1):
      el=x[i:j]
      distinct=set(el)
      if len(distinct)<=y and len(distinct)==z:  
        print(x[i:j],end=',')
  print(" ")


def substring_distchar(x,z):
  print("All possible substrings with",z,"distinct characters:")
  l=len(x)
  for i in range(0,l+1):
    for j in range(i+1,l+1):
      el=x[i:j]
      distinct=set(el)
      if len(distinct)==z:  
        print(x[i:j],end=',')
  print(" ")


def palindrome_check(x):
  print("Palindrome substrings:")
  l=len(x)
  for i in range(0,l+1):
    for j in range(i+1,l+1):
      x1=x[i:j]
      x2=x1[::-1]
      if (x1==x2):
        print(x1, end=",")
  #print(" ")



string=str(input("Enter a string= "))
substrings(string)

length=int(input("Enter the preferred length of substring= "))
substring_with_length(string,length)

dist=int(input("Enter how many distinct characters you prefer="))
substring_length_distchar(string,length,dist)

substring_distchar(string,dist)

palindrome_check(string)

```
